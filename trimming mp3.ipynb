```python
from mutagen.mp3 import MP3
import os
from pydub import AudioSegment
import soundfile as sf
import librosa
```

```python
def get_len_marah(file=None):
    folder = "1000data/marah/"
    if file == None:
        for filename in os.listdir(folder):
            if filename.endswith(".mp3"): 
                audio = MP3(folder+filename)
                global list_marah
                list_marah.append(audio.info.length)
                continue
            else:
                continue
    else: 
        audio = MP3(file)
        return audio.info.length
```

# Set second to trims

```python
second = 4.0
```

```python
def trim_senang():
#     second = 7.0
    for filename in os.listdir("1000data/senang/"):
        if filename.endswith(".mp3"): 
            durasi = get_len_senang("1000data/senang/"+filename)
            if durasi > second :
#                 print("./1000data/senang/"+filename,":",durasi)
                song = AudioSegment.from_mp3("1000data/senang/"+filename)
                seconds = second * 1000
                trimmed = song[:seconds]
                trimmed.export("data_trim_4s/senang/"+filename+"_02.mp3", format="mp3")
            continue
        else:
            continue
```
